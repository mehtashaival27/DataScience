# Matplotlib

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(10,20)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot')
plt.savefig('Scatter.png')
```

```python
plt.plot(x,y,'b*--')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b')
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'r--')
plt.subplot(2,2,4)
plt.plot(x,y,'g*--')
```

```python
# Sin Wave
a = np.arange(0, 5 * np.pi, 0.1)
b = np.sin(a)
plt.title('Sin Wave')
plt.plot(a,b)
```

```python
# Bar Chart
b1 = [2,3,4,5,6]
b2 = [4,8,12,16,20]
plt.bar(b1,b2,3)
```

```python
# Histograms
plt.hist(b1)
```

```python
# Box Plot using Matplotlib
results = [np.random.normal(0,std,100) for std in range(1,5)]
plt.boxplot(results, patch_artist=True, vert=True)
```
